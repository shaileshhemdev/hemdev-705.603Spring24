If you see the above this dataset does not have any missing values. So we don't have any corrective action to take. We will proceed with the derivation of features and then dropping the ones we said we would

```python
from datetime import datetime, date 

# Let's convert transaction date and time and dob to date-time
df["dob_dt"] = pd.to_datetime(df['dob'], format='%Y-%m-%d', errors='coerce')
df["txn_dt"] = pd.to_datetime(df['trans_date'], format='%Y-%m-%d', errors='coerce')
df["txn_time"] = pd.to_datetime(df['trans_time'], format='%H:%M:%S', errors='coerce')

# Compute weekday from transaction date
df['txn_weekday'] = df['txn_dt'].dt.day_name()

# This function converts given date to age 
def age(born): 
    born = datetime.strptime(born, "%Y-%m-%d").date() 
    today = date.today() 
    return today.year - born.year - ((today.month,  today.day) < (born.month,  born.day)) 

# Compute age from date of birth 
df['age'] = df['dob'].apply(age) 

# Compute hour from transaction date
df['txn_hour'] = df['txn_time'].dt.hour

# Slot the times into well known time ranges
time_ranges = [0,4,8,12,16,21,24]
part_of_day_dict = ['Late Night', 'Early Morning','Morning','Afternoon','Evening','Night']
df['part_of_day'] = pd.cut(df['txn_hour'], bins=time_ranges, labels=part_of_day_dict, include_lowest=True)

# Compute month from transaction date
df['txn_month'] = df['txn_dt'].dt.month_name()

def haversine_vectorize(lon1, lat1, lon2, lat2):
    """Returns distance, in kilometers, between one set of longitude/latitude coordinates and another"""
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
 
    newlon = lon2 - lon1
    newlat = lat2 - lat1
 
    haver_formula = np.sin(newlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(newlon/2.0)**2
 
    dist = 2 * np.arcsin(np.sqrt(haver_formula ))
    miles = 3958 * dist #6367 for distance in KM 
    return miles

def classify_distance(_haversine_distance): 
    if (int(_haversine_distance) < 20):
        return 'Low'
    elif (int(_haversine_distance) >= 50):
        return 'High'
    else:
        return 'Medium'

df['distance_from_merchant'] = haversine_vectorize(df['lat'],df['long'],df['merch_lat'],df['merch_long'])
df['txn_dist'] = df['distance_from_merchant'].apply(classify_distance) 

df.head()
```

#### Deep Learning Model

With 2 traditional models tried, now lets build the LSTM Network using Pytorch. We will need the following components

<ul>
    <li>
        <b>Dataset:</b> This will take the training and test data and using a supplied lookback parameter construct a tensor of current input and all the previous N inputs where is N is the lookback. E.g. If we say lookback = 7 it will look at previous 7 values
    </li>
    <li>
        <b>Neural Network Model:</b> Class encapsulating our Model which in turn uses Pytorch NN Module
    </li>
    <li>
        <b>Parameters:</b> We tried different parameter values for input layer (1, 8, 52), batch_size (1, 17) balancing accuracy with training time. 
    </li>
</ul>

```python
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as data
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Initialize a MinMaxScaler
scaler = MinMaxScaler()

# Parameters for the model
INPUT_LAYER_SIZE = 52
HIDDEN_LAYER_SIZE = 50
NN_LAYERS = 1
OUTPUT_LAYER_SIZE = 1
LOOKBACK_PARAMETER = 14
BATCH_PARAMETER = 8
SHUFFLE_PARAMETER = True
EPOCH_PARAMETER = 2000


def create_dataset(dataset, lookback):
    """Transform a time series into a prediction dataset
    
    Args:
        dataset: A numpy array of time series, first dimension is the time steps
        lookback: Size of window for prediction
    """
    X, y = [], []
    for i in range(len(dataset)-lookback):
        feature = dataset[i:i+lookback]
        target = dataset[i+1:i+lookback+1]
        X.append(feature)
        y.append(target)
    return torch.tensor(X), torch.tensor(y)

class FraudNNModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(input_size=INPUT_LAYER_SIZE, hidden_size=HIDDEN_LAYER_SIZE, num_layers=NN_LAYERS, batch_first=True)
        self.linear = nn.Linear(HIDDEN_LAYER_SIZE, OUTPUT_LAYER_SIZE)
    def forward(self, x):
        x, _ = self.lstm(x)
        x = self.linear(x)
        return x

def plot_predictions_vs_actuals(model, timeseries, train_size, X_train, X_test):
    with torch.no_grad():
        # Shift train predictions for plotting
        train_plot = np.ones_like(timeseries) * np.nan
        
        y_pred = model(X_train)
        y_pred = y_pred[:, -1, :]
        train_plot[LOOKBACK_PARAMETER:train_size] = model(X_train)[:, -1, :]
        
        # Shift test predictions for plotting
        test_plot = np.ones_like(timeseries) * np.nan
        test_plot[train_size+LOOKBACK_PARAMETER:len(timeseries)] = model(X_test)[:, -1, :]
    
    # Plot the progress
    plt.plot(timeseries, c='b')
    plt.plot(train_plot, c='r')
    plt.plot(test_plot, c='g')
    plt.show()
```

We will now apply the LSTM Neural Network to Total Transactions

```python
# Create the Training and Test Data
pytorch_train = train[["y"]].values.astype('float32')
pytorch_test = test[["y"]].values.astype('float32')

# Apply scaling to let the network learn effectively
scaler.fit(pytorch_train)
scaled_train = scaler.transform(pytorch_train)
scaled_test = scaler.transform(pytorch_test)

# Ensure the overall data is also converted and scaled as the training and test 
timeseries = total_txn_agg_df[["y"]].values.astype('float32')
timeseries =  scaler.transform(timeseries)

# Create the dataset converting our values to Tensors
X_train, y_train = create_dataset(scaled_train, lookback=LOOKBACK_PARAMETER)
X_test, y_test = create_dataset(scaled_test, lookback=LOOKBACK_PARAMETER)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

# Plot the data
plt.plot(timeseries)
plt.show()
```

```python
# Initialize FraudNNModel for Total Transactions
model = FraudNNModel()
optimizer = optim.Adam(model.parameters())
loss_fn = nn.MSELoss()
loader = data.DataLoader(data.TensorDataset(X_train, y_train), shuffle=True, batch_size=BATCH_PARAMETER)

n_epochs = EPOCH_PARAMETER
for epoch in range(n_epochs):
    model.train()
    for X_batch, y_batch in loader:
        y_pred = model(X_batch)
        loss = loss_fn(y_pred, y_batch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    # Validation
    if epoch % 100 != 0:
        continue
    model.eval()
    with torch.no_grad():
        y_pred = model(X_train)
        train_rmse = np.sqrt(loss_fn(y_pred, y_train))
        y_pred = model(X_test)
        test_rmse = np.sqrt(loss_fn(y_pred, y_test))
    print("Epoch %d: train RMSE %.4f, test RMSE %.4f" % (epoch, train_rmse, test_rmse))
```

```python
plot_predictions_vs_actuals(model, timeseries, 209, X_train, X_test)
```

Based on the above we can conclude that the Deep Learning Model is performing really well for all transactions compared to traditional model since the RMSE is really low both for Train and Test
